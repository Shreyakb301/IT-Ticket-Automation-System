# IT Ticket Dataset EDA
Explore class balance and ticket text patterns before training.

```
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

df = pd.read_csv(Path("../data/tickets.csv"))
df.head()
```

```
df.shape, df.columns.tolist()
```

```
df[["category", "subcategory", "priority", "ticket_text"]].sample(10)
```

```
df["category"].value_counts().plot(kind="bar", title="Tickets by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```
df["priority"].value_counts().plot(kind="bar", title="Tickets by Priority")
plt.xlabel("Priority")
plt.ylabel("Count")
plt.show()
```
